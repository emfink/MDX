```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import matplotlib.cm as cm

# 1. Create the original graph
G1 = nx.petersen_graph()

# 2. Assign unique colors to nodes
colors = [cm.rainbow(i / len(G1.nodes())) for i in range(len(G1.nodes()))]

# 3. Scramble it
nodes = list(G1.nodes())
scrambled_nodes = random.sample(nodes, len(nodes))
mapping = dict(zip(nodes, scrambled_nodes))
G2 = nx.relabel_nodes(G1, mapping)

color_map_G2 = [None] * len(G2.nodes())
for original_node, scrambled_node in mapping.items():
    color_map_G2[list(G2.nodes()).index(scrambled_node)] = colors[original_node]

# 4. Visualization with Transparency
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Set figure and axes background to transparent
fig.patch.set_alpha(0.0)
ax1.patch.set_alpha(0.0)
ax2.patch.set_alpha(0.0)

# Plot 1: Standard Circular Layout
pos1 = nx.circular_layout(G1)
nx.draw(G1, pos1, node_color=colors, with_labels=False, node_size=1000, ax=ax1)
ax1.set_title("Identity (Original Structure)", fontsize=14, color='white') # Set text to white if background is dark

# Plot 2: A chaotic 'Spring' Layout
pos2 = nx.spring_layout(G2, seed=42)
nx.draw(G2, pos2, node_color=color_map_G2, with_labels=False, node_size=1000, ax=ax2)
ax2.set_title("Scrambled (Same Connections, Different View)", fontsize=14, color='white')

plt.tight_layout()

# Save with transparent background
plt.savefig("petersen_graph.png", transparent=True, dpi=300)
plt.show()

print(f"Mathematical check: Are they the same? {nx.is_isomorphic(G1, G2)}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import matplotlib.cm as cm

# 1. Create the graph
G1 = nx.petersen_graph()

# 2. Assign high-contrast colors
node_colors = [cm.rainbow(i / len(G1.nodes())) for i in range(len(G1.nodes()))]

# 3. Scramble it
nodes = list(G1.nodes())
scrambled_nodes = random.sample(nodes, len(nodes))
mapping = dict(zip(nodes, scrambled_nodes))
G2 = nx.relabel_nodes(G1, mapping)

# Map the colors to the new node labels
color_map_G2 = [None] * len(G2.nodes())
for original_node, scrambled_node in mapping.items():
    color_map_G2[list(G2.nodes()).index(scrambled_node)] = node_colors[original_node]

# 4. Visualization with "Mega-Lines"
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7), facecolor='white')

# Style settings for maximum visibility
style_settings = {
    'node_size': 1200,        # Jumbo dots
    'linewidths': 0,          # No borders
    'width': 6.0,             # Ultra-thick lines
    'edge_color': '#333333',  # High-contrast dark grey
    'alpha': 0.8,             # Slight transparency for better intersections
    'with_labels': False
}

# Plot 1: Orderly
pos1 = nx.circular_layout(G1)
nx.draw(G1, pos1, node_color=node_colors, ax=ax1, **style_settings)
ax1.set_title("Clean Structure", fontsize=22, fontweight='bold', pad=20)

# Plot 2: Chaotic
pos2 = nx.spring_layout(G2, k=2.0, seed=42)
nx.draw(G2, pos2, node_color=color_map_G2, ax=ax2, **style_settings)
ax2.set_title("Scrambled Logic", fontsize=22, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()
```

```python
import graphviz

# 1. Initialize the diagram
# 'rankdir=BT' makes the arrows point 'Bottom to Top' (standard for inheritance)
dot = graphviz.Digraph('Vehicle_Hierarchy', graph_attr={'rankdir': 'BT', 'size': '10'})

# 2. Global Styles for the "Bubbles"
dot.attr('node', shape='ellipse', style='filled', fillcolor='#EBF5FB', color='#2E86C1', fontname='Arial')

# 3. Create Class Nodes
# Format: dot.node(ID, Label)
dot.node('V', 'Vehicle\n—\nVIN\nmanufacturer', fillcolor='#F7DC6F')
dot.node('MV', 'MotorizedVehicle\n—\nengineType\nfuelLevel')
dot.node('B', 'Bicycle\n—\ngearCount\nframeType')
dot.node('ST', 'SemiTruck\n—\ntrailerType\nmaxTowing')
dot.node('ES', 'ElectricSedan\n—\nbatteryKWh\nsoftwareVer')
dot.node('CS', 'CargoShip\n—\ndeadweightTonnage')
dot.node('D', 'Drone\n—\nrotorCount\nsignalRange', fillcolor='#EAEDED')

# 4. Create Interface Nodes (Rectangles)
dot.attr('node', shape='box', style='filled', fillcolor='#D5F5E3', color='#1D8348')
dot.node('I_RR', '<<Interface>>\nIFuelRefillable')
dot.node('I_PP', '<<Interface>>\nIPedalPowered')
dot.node('I_CC', '<<Interface>>\nICargoCarrier')
dot.node('I_L', '<<Interface>>\nILoggable')
dot.node('I_C', '<<Interface>>\nIChargeable')
dot.node('I_N', '<<Interface>>\nINavigation')
dot.node('I_RC', '<<Interface>>\nIRemoteControlled')

# 5. Connect Inheritance (Solid Lines)
dot.edge('MV', 'V')
dot.edge('B', 'V')
dot.edge('CS', 'V')
dot.edge('ST', 'MV')
dot.edge('ES', 'MV')

# 6. Connect Interfaces (Dashed Lines)
dot.attr('edge', style='dashed', color='#CB4335')
dot.edge('MV', 'I_RR')
dot.edge('B', 'I_PP')
dot.edge('ST', 'I_CC')
dot.edge('ST', 'I_L')
dot.edge('ES', 'I_C')
dot.edge('CS', 'I_CC')
dot.edge('CS', 'I_N')
dot.edge('D', 'I_CC')
dot.edge('D', 'I_RC')

# 7. DISPLAY IN COLAB
dot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Data ---
years = [2020, 2022, 2024, 2026]
areas = ['Protein/Drug\nDiscovery', 'Medical\nImaging', 'Physics\nSimulations', 'Climate/\nWeather', 'Math\nProofs']

# Data reflects Medical Imaging's leadership in the field
data = {
    2020: [2000, 3500, 400, 250, 50],
    2022: [8000, 11000, 1200, 900, 150],
    2024: [25000, 24000, 8000, 5500, 1200],
    2026: [50000, 42000, 22000, 14000, 6000]
}

# --- Styling ---
area_colors = ['#8200FF', '#00FF87', '#FF007F', '#00D4FF', '#FFB800']
AI_DARK = '#381E69'
MAX_VAL = 55000

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(24, 12), sharey=True)
fig.patch.set_facecolor('white')

for i, year in enumerate(years):
    ax = axes[i]
    values = data[year]
    x_pos = np.arange(len(areas))

    bars = ax.bar(x_pos, values, color=area_colors, width=0.8, alpha=0.9, edgecolor=AI_DARK)
    ax.set_ylim(0, MAX_VAL)

    # BIG Fonts for presentation visibility
    ax.set_title(f"{year}", fontsize=36, fontweight='bold', color=AI_DARK, pad=30)

    if i == 0:
        ax.tick_params(axis='y', labelsize=28, colors=AI_DARK)
    else:
        ax.tick_params(axis='y', length=0)

    # Rotated labels on top of bars
    for bar in bars:
        val = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, val + 1000,
                f'{int(val):,}', ha='center', va='bottom',
                fontsize=20, fontweight='bold', color=AI_DARK, rotation=90)

    ax.set_xticks(x_pos)
    ax.set_xticklabels(areas, fontsize=16, fontweight='bold', color=AI_DARK, rotation = 45)

    for spine in ['top', 'right', 'left']:
        ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.savefig('ai_science_comprehensive_5cat.png', dpi=300)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Data & Settings ---
years = [2020, 2022, 2024, 2026]
total_ai = [140000, 195000, 242000, 310000]
ai_science = [1900, 6750, 29700, 77000]
percentages = [(sci / tot) * 100 for sci, tot in zip(ai_science, total_ai)]

AI_PURPLE = '#8200FF'
SCIENCE_TEAL = '#00D4FF'
AI_DARK = '#381E69'

fig, ax = plt.subplots(figsize=(18, 12))
x = np.arange(len(years))
width = 0.6

# 1. Total AI Bar (The Background/Container)
ax.bar(x, total_ai, width, label='Total AI Publications',
       color=AI_PURPLE, alpha=0.3, edgecolor=AI_PURPLE, linewidth=2)

# 2. Science Bar (The "Inside" Subset)
bars_sci = ax.bar(x, ai_science, width, label='AI for Science (Subset)',
                 color=SCIENCE_TEAL, alpha=1.0, edgecolor=AI_DARK, linewidth=2)

# --- Labels ---
for i, bar in enumerate(bars_sci):
    total_h = total_ai[i]
    sci_h = bar.get_height()

    # Label for Total (at top of purple)
    ax.text(bar.get_x() + bar.get_width()/2., total_h + 5000,
            f'Total: {total_h:,}', ha='center', fontsize=22, fontweight='bold', color=AI_PURPLE)

    # Label for Science (inside or above teal)
    y_pos = sci_h / 2 if i >= 2 else sci_h + 3000
    color = 'white' if i >= 2 else AI_DARK
    ax.text(bar.get_x() + bar.get_width()/2., y_pos,
            f'{sci_h:,}\n({percentages[i]:.1f}%)',
            ha='center', va='center', fontsize=20, fontweight='black', color=color)

# BIG Fonts
ax.set_ylim(0, 380000)
ax.set_title('AI for Science as a Subset of Global AI Research', fontsize=36, fontweight='bold', pad=50)
ax.set_xticklabels(years, fontsize=28, fontweight='bold')
ax.tick_params(axis='y', labelsize=26)
ax.legend(fontsize=22, loc='upper left', frameon=False)

plt.tight_layout()
plt.savefig('ai_science_nested_bars.png', dpi=300)
```
